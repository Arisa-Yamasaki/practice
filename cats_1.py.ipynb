```python
train = []
for i,j in cat_data['train'].items():
    for r in j:
        train.append((r.ravel()/255, 1 if i=='cat' else 0))

# i: cat or not_cat
# j: image
# r: pixels in 64*64「この8ビット画像では、濃淡を2の8乗＝256階調まで表すことができます。画素値0が黒、画素値255が白になります。」
#(())にしないと一つだと認識されないからerrorを吐く
```
